```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
x = 5 * np.random.rand(500,1)
y = 0.5 * x + 0.9 + 0.2 * np.random.randn(500,1)
plt.scatter(x,y)
plt.show()
```

```python
class liner():
    
    def _init_(self):
        self.w = None
        self.b = None
        
    def loss(self,x,y):
        num_train = x.shape[0]
        pre = x.dot(self.w)   + self.b
        loss = np.sum(np.square(pre - y)) / num_train
        dw = (np.sum(x*(pre - y))) / num_train 
        db = np.sum(pre - y) / num_train 
        return loss,dw,db
    
    def train(self,x,y,learn_rate = 0.0001 ,iters = 10000):
        num_feature = x.shape[1]
        self.w = np.zeros((num_feature,1))
        self.b = 0
        loss_list = []
        for i in range(iters):
            loss,dw,db = self.loss(x ,y)
            loss_list.append(loss)
            self.w += -learn_rate * dw
            self.b += -learn_rate * db
            if i%1000 == 0:
                print("iters = %d , loss = %f"%(i,loss))
         
        return loss_list   
        
```

```python
new = liner()
loss_list = new.train(x,y)
```

```python
import matplotlib.pyplot as plt
f = x.dot(new.w) + new.b
fig = plt.figure()
plt.subplot(211)
plt.scatter(x,y,color = 'black')
plt.plot(x,f,color = 'red')
plt.xlabel('X')
plt.ylabel('y')
plt.subplot(212)
plt.plot(loss_list,color = 'blue')
plt.xlabel('epochs')
plt.ylabel('errors')
plt.show()
```
